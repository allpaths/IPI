This notebook follows a lecture on Claude Shannon's information theory, originally developed and published in **1948**. The central goal is to derive the formula for **Shannon Information Entropy** using a rigorous axiomatic approach.

Shannon's problem was to define a mathematical way of **quantifying the amount of information** extracted from observing an event. His framework applies to any process with a probabilistic nature — coin tosses, characters in a message, genomic sequences, physical events, and more.

## Section 1: Shannon's Axiomatic Approach

Rather than guessing a formula, Shannon derived his theory from first principles using four **axioms** — fundamental properties that any reasonable measure of information *must* satisfy.

Let $I(p)$ denote the **information content** extracted from observing an event that has probability $p$ of occurring.

### Axiom 1 — Information is a function of probability
$$I = I(p)$$
The amount of information gained from an event depends only on its probability. More surprising events (lower $p$) should carry more information.

### Axiom 2 — Information is non-negative and continuous
$$I(p) \geq 0$$
$I(p)$ is a continuous function of $p$, meaning it can be differentiated with respect to $p$.

### Axiom 3 — A certain event carries no information
$$p = 1 \implies I(1) = 0$$
If an event is guaranteed to happen (probability = 1), observing it tells you nothing new. There is no uncertainty to resolve.

### Axiom 4 — Information is additive for independent events
For $N$ independent events with individual probabilities $p_1, p_2, \ldots, p_N$:
- The **joint probability** is the product: $p = p_1 \cdot p_2 \cdot p_3 \cdots p_N$
- The **total information** is the sum of individual informations:
$$I(p) = I(p_1 \cdot p_2 \cdots p_N) = I(p_1) + I(p_2) + \cdots + I(p_N)$$

This additivity axiom is crucial — it links multiplication of probabilities to addition of information.

## Section 2: Identifying the Information Function

Looking at the four axioms carefully, we need a function $I(p)$ that:
1. Depends only on $p$
2. Is non-negative and continuous
3. Equals zero when $p = 1$
4. Converts products into sums (i.e., $I(p_1 \cdot p_2) = I(p_1) + I(p_2)$)

**The only class of functions satisfying all four axioms is the logarithm.**

Specifically, for an event with probability $p$:
$$I(p) = -\log_b(p) = \log_b\left(\frac{1}{p}\right)$$

The **minus sign** is required because:
- Probabilities are in the range $(0, 1]$, so $\log_b(p) \leq 0$ for $b > 1$
- Axiom 2 requires $I(p) \geq 0$
- Negating gives us the correct sign

These are all equivalent ways to write the same formula:
$$I(p) = -\log_b(p) = \log_b\left(p^{-1}\right) = \log_b\left(\frac{1}{p}\right)$$

**Verification of Axiom 4:**
$$I(p_1 \cdot p_2) = \log_b\left(\frac{1}{p_1 p_2}\right) = \log_b\left(\frac{1}{p_1}\right) + \log_b\left(\frac{1}{p_2}\right) = I(p_1) + I(p_2) \checkmark$$

## Section 3: Units of Information and Base Change

### The role of the base $b$

The base $b$ of the logarithm determines the **units** of information. The base must satisfy $b > 1$.

| Base $b$ | Unit of information |
|----------|--------------------|
| $b = 2$  | **bits** (binary digits) |
| $b = 3$  | **trits** (ternary digits) |
| $b = e$  | **nats** (natural units) |

The **bit** is the standard unit of digital information — and it arises directly from this theoretical framework with base 2.

### Changing the base

The choice of base is largely a matter of convention, because bases can be freely converted. The **change-of-base formula** is:

$$\log_b\left(\frac{1}{p}\right) = \frac{\log_a\left(\frac{1}{p}\right)}{\log_a(b)}$$

### Derivation of the base-change formula

Let $x = \log_b\left(\frac{1}{p}\right)$. By definition of logarithm:
$$\frac{1}{p} = b^x$$

Apply $\log_a$ to both sides:
$$\log_a\left(\frac{1}{p}\right) = \log_a(b^x) = x \cdot \log_a(b)$$

Substituting back $x = \log_b\left(\frac{1}{p}\right)$:
$$\log_a\left(\frac{1}{p}\right) = \log_b\left(\frac{1}{p}\right) \cdot \log_a(b)$$

Solving for the original expression:
$$\boxed{\log_b\left(\frac{1}{p}\right) = \frac{\log_a\left(\frac{1}{p}\right)}{\log_a(b)}}$$

### Example: converting nats to bits

Starting from $b = e$ (nats), convert to $a = 2$ (bits):
$$I(p) = \ln\left(\frac{1}{p}\right) \text{ [nats]} = \ln(2) \cdot \log_2\left(\frac{1}{p}\right) \text{ [bits]}$$

where $\ln(x) = \log_e(x)$ is the natural logarithm.

```python
import numpy as np

def information(p, base=2):
    """Compute information content I(p) = log_b(1/p) in the given base."""
    if p <= 0 or p > 1:
        raise ValueError("Probability must be in (0, 1]")
    return np.log(1/p) / np.log(base)

# Demonstrate that all bases give equivalent information (just different units)
p = 0.25  # probability of 25%

I_bits = information(p, base=2)
I_nats = information(p, base=np.e)
I_trits = information(p, base=3)

print(f"Information for p = {p}:")
print(f"  I = {I_bits:.4f} bits")
print(f"  I = {I_nats:.4f} nats")
print(f"  I = {I_trits:.4f} trits")

# Verify base conversion: nats -> bits using the formula
I_bits_from_nats = I_nats / np.log(2)  # divide by ln(2) = log_e(2)
print(f"\nConverting nats to bits: {I_nats:.4f} / ln(2) = {I_bits_from_nats:.4f} bits")
print(f"Direct bits calculation:  {I_bits:.4f} bits")
print(f"Match: {np.isclose(I_bits, I_bits_from_nats)}")

# Show that a certain event (p=1) carries zero information
print(f"\nInformation for p=1 (certain event): {information(1.0, base=2):.4f} bits")

# Show that a rare event carries more information
print("\nInformation vs. probability:")
for p_val in [1.0, 0.5, 0.25, 0.1, 0.01]:
    print(f"  p = {p_val:.2f}  -->  I = {information(p_val, base=2):.4f} bits")
```

## Section 4: Setting Up Shannon Information Entropy

So far we have a formula for the information gained from a **single event** with probability $p$. Now we ask:

> **How much information is contained in a message of $N$ characters (or equivalently, in a set of $N$ events)?**

### Defining N and U

Consider a message (or sequence of events):
- $N$ = **total number** of characters/events in the set
- The characters are: $x_1, x_2, x_3, \ldots, x_N$ (can include repeats)

Some characters may appear multiple times. We define:
- $U$ = **number of distinct** characters/events, where $1 \leq U \leq N$
- $\mathcal{X} = \{x_1, x_2, \ldots, x_U\}$ = the set of **unique** characters
- $H(\mathcal{X})$ = **information entropy** of this set

### Example

Consider the message: $\{A, A, T, Q, M\}$
- $N = 5$ (five total characters)
- Unique characters: $\mathcal{X} = \{A, T, Q, M\}$
- $U = 4$ (four distinct characters)

### Probability distribution

For each distinct character $x_j \in \mathcal{X}$, assign a probability $p_j$:
$$\mathcal{P} = \{p_1, p_2, \ldots, p_U\}$$

These must satisfy:
1. **Non-negativity:** $p_j \geq 0$ for all $j$
2. **Normalization:** $\displaystyle\sum_{j=1}^{U} p_j = 1$

## Section 5: Deriving Shannon Information Entropy $H(\mathcal{X})$

We want to find the **average information per character** when observing a long message.

### Step 1 — Information from observing character $x_j$

When we observe character $x_j$ (which has probability $p_j$), we gain:
$$I(x_j) = \log_b\left(\frac{1}{p_j}\right)$$

### Step 2 — How many times does $x_j$ appear in $N$ observations?

Over $N$ independent observations, character $x_j$ appears on average:
$$\text{occurrences of } x_j = N \cdot p_j$$

### Step 3 — Total information from all observations

Summing over all $U$ distinct characters:
$$I_{\text{total}} = \sum_{j=1}^{U} \underbrace{(N \cdot p_j)}_{\text{occurrences}} \cdot \underbrace{\log_b\left(\frac{1}{p_j}\right)}_{\text{info per occurrence}}$$

### Step 4 — Average information per character

Divide by $N$:
$$\frac{I_{\text{total}}}{N} = \frac{1}{N} \sum_{j=1}^{U} N \cdot p_j \cdot \log_b\left(\frac{1}{p_j}\right)$$

The $N$ cancels:
$$\frac{I_{\text{total}}}{N} = \sum_{j=1}^{U} p_j \cdot \log_b\left(\frac{1}{p_j}\right)$$

### Step 5 — Final form: Shannon Information Entropy

This quantity is the **Shannon information entropy** $H(\mathcal{X})$. Writing $\log_b(1/p_j) = -\log_b(p_j)$:

$$\boxed{H(\mathcal{X}) = -\sum_{j=1}^{U} p_j \log_b(p_j)}$$

This is the **information content per character** (or per event) for a source described by the probability distribution $\{p_1, \ldots, p_U\}$.

```python
import numpy as np

def shannon_entropy(probabilities, base=2):
    """
    Compute Shannon information entropy H(X) for a discrete distribution.
    
    H(X) = -sum(p_j * log_b(p_j))
    
    Parameters
    ----------
    probabilities : array-like of floats in (0, 1]
    base : logarithm base (2=bits, e=nats, 3=trits)
    
    Returns
    -------
    float: entropy in units determined by base
    """
    probs = np.array(probabilities, dtype=float)
    # Filter out zero probabilities (limit p*log(p) -> 0 as p -> 0)
    probs = probs[probs > 0]
    assert np.isclose(probs.sum(), 1.0), "Probabilities must sum to 1"
    return -np.sum(probs * np.log(probs) / np.log(base))

# Example: fair coin (two equally likely outcomes)
H_coin = shannon_entropy([0.5, 0.5], base=2)
print(f"Fair coin  [p=0.5, p=0.5]:        H = {H_coin:.4f} bits")

# Example: biased coin
H_biased = shannon_entropy([0.9, 0.1], base=2)
print(f"Biased coin [p=0.9, p=0.1]:       H = {H_biased:.4f} bits")

# Example: uniform over 4 symbols (like our A,T,Q,M example)
H_4 = shannon_entropy([0.25, 0.25, 0.25, 0.25], base=2)
print(f"4 equal symbols [0.25 each]:      H = {H_4:.4f} bits")

# Example: non-uniform over 4 symbols
H_4_nonuniform = shannon_entropy([0.5, 0.25, 0.125, 0.125], base=2)
print(f"4 unequal symbols [0.5,0.25,...]: H = {H_4_nonuniform:.4f} bits")

# Total information for a message of N=100 characters
N = 100
I_total = N * H_4
print(f"\nTotal information for N={N} characters with H={H_4:.4f}: I_total = {I_total:.2f} bits")
```

## Section 6: Boundary Cases

### Case 1: Certain event ($p_j = 1$)

If one event is certain, then $\log_b(1) = 0$, so:
$$p_j = 1 \implies H(\mathcal{X}) = 0$$

**Interpretation:** A process with no uncertainty contains no information. You learn nothing from observing something you already know with 100% certainty.

### Case 2: Impossible event ($p_j = 0$)

Here we cannot directly compute $\log_b(0)$ — it is undefined (approaches $-\infty$).

However, we can take the limit:
$$\lim_{p \to 0^+} p \cdot \log_b\left(\frac{1}{p}\right) = 0$$

This is a well-known result (application of L'Hôpital's rule). Therefore:
$$p_j \to 0 \implies p_j \log_b\left(\frac{1}{p_j}\right) \to 0$$

**Interpretation:** An impossible event also contributes nothing to the entropy — it never occurs, so it never provides information.

This means terms with $p_j = 0$ can be safely excluded from the entropy sum, which is why the convention $0 \log 0 = 0$ is used.

```python
import numpy as np
import matplotlib.pyplot as plt

# Demonstrate the limit: p * log(1/p) -> 0 as p -> 0
p_values = np.logspace(-10, 0, 1000)  # from 1e-10 to 1
term = p_values * np.log2(1.0 / p_values)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Left plot: p * log(1/p) showing it goes to 0 at both ends
axes[0].semilogx(p_values, term, 'steelblue', linewidth=2)
axes[0].set_xlabel('Probability $p$', fontsize=12)
axes[0].set_ylabel('$p \\cdot \\log_2(1/p)$', fontsize=12)
axes[0].set_title('Entropy term $p \\log(1/p)$\nvanishes at both boundaries', fontsize=11)
axes[0].axhline(0, color='gray', linewidth=0.8, linestyle='--')
axes[0].grid(True, alpha=0.3)
axes[0].annotate('p→0: term→0\n(L\'Hôpital)', xy=(1e-8, 0.005), fontsize=9, color='darkred')
axes[0].annotate('p=1: term=0\n(log 1 = 0)', xy=(0.8, 0.005), fontsize=9, color='darkred')

# Right plot: I(p) = log2(1/p)
p_vals2 = np.linspace(0.001, 1.0, 500)
I_vals = np.log2(1.0 / p_vals2)
axes[1].plot(p_vals2, I_vals, 'darkorange', linewidth=2)
axes[1].set_xlabel('Probability $p$', fontsize=12)
axes[1].set_ylabel('$I(p) = \\log_2(1/p)$ [bits]', fontsize=12)
axes[1].set_title('Information Content $I(p)$\nRare events → more information', fontsize=11)
axes[1].axhline(0, color='gray', linewidth=0.8, linestyle='--')
axes[1].grid(True, alpha=0.3)
axes[1].annotate('p=1 → I=0\n(certain)', xy=(0.85, 0.3), fontsize=9, color='darkred')
axes[1].annotate('p→0 → I→∞\n(impossible)', xy=(0.05, 8), fontsize=9, color='darkred')

plt.tight_layout()
plt.savefig('information_content.png', dpi=100, bbox_inches='tight')
plt.show()
print("Figure saved.")
```

## Section 7: Total Information

The Shannon entropy $H(\mathcal{X})$ gives the **information per character** (or per event). For a full message of $N$ characters:

$$\boxed{I_{\text{total}} = N \cdot H(\mathcal{X})}$$

This is the total expected information content when observing all $N$ characters.

---

## Section 8: Connection to Thermodynamic Entropy

Shannon information entropy closely resembles the **Boltzmann entropy** from statistical thermodynamics:

$$S = -k_B \sum_{j} p_j \ln(p_j)$$

where $k_B$ is the Boltzmann constant and the logarithm is the natural log (giving units of J/K).

Comparing with Shannon's formula:
$$H(\mathcal{X}) = -\sum_{j=1}^{U} p_j \log_b(p_j)$$

The two differ only in:
- The **Boltzmann constant** $k_B$ (makes thermodynamic entropy have physical units)
- The **base** of the logarithm (natural log vs. log base 2)

This deep structural similarity is why Shannon called his measure **information entropy**. However, $H(\mathcal{X})$ has units of **bits**, not joules per kelvin, and describes information, not heat.

---

## Section 9: Continuous Shannon Entropy

The formula derived so far applies to **discrete** probability distributions (finite or countable outcomes).

For a **continuous** random variable $X$ with probability density function $p(x)$, the entropy becomes an **integral** instead of a sum:

$$\boxed{H(X) = -\int p(x) \cdot \log_b\bigl(p(x)\bigr)\, dx}$$

This is called the **differential entropy** of the continuous distribution $p(x)$.

Note: differential entropy has some different properties from discrete entropy (it can be negative), but it plays the same conceptual role in quantifying uncertainty.

## Section 10: Maximum Shannon Entropy

### When is entropy maximized?

A natural question: for a given set of $U$ distinct symbols, what probability distribution $\{p_1, \ldots, p_U\}$ **maximizes** $H(\mathcal{X})$?

**Answer:** The entropy is maximized when all probabilities are **equal**:
$$p_j = \frac{1}{U} \quad \text{for all } j$$

### Deriving the maximum value

Substituting $p_j = 1/U$ into the entropy formula:

$$H(\mathcal{X}) = -\sum_{j=1}^{U} \frac{1}{U} \log_b\left(\frac{1}{U}\right)$$

Since $\frac{1}{U}\log_b(\frac{1}{U})$ does not depend on $j$, the sum just multiplies it by $U$:

$$H(\mathcal{X}) = -\frac{1}{U} \cdot U \cdot \log_b\left(\frac{1}{U}\right) = -\log_b\left(\frac{1}{U}\right) \cdot \frac{U}{U}$$

$$\boxed{H(\mathcal{X})_{\max} = \log_b(U)} \qquad \text{when } p_j = \frac{1}{U} \text{ for all } j$$

### Physical interpretation

Maximum entropy occurs when the distribution is **maximally uncertain** — no symbol is more likely than any other. This corresponds to the most "scrambled" or disordered message. Any bias toward certain symbols (unequal probabilities) reduces the entropy.

```python
import numpy as np
import matplotlib.pyplot as plt

# Visualization: entropy of a binary source as a function of p
# H(p) = -p*log2(p) - (1-p)*log2(1-p)

p = np.linspace(0.001, 0.999, 1000)
H = -(p * np.log2(p) + (1-p) * np.log2(1-p))

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# --- Left: binary entropy curve ---
axes[0].plot(p, H, 'steelblue', linewidth=2.5)
axes[0].axvline(0.5, color='crimson', linestyle='--', linewidth=1.5, label='p=0.5 (max entropy)')
axes[0].axhline(1.0, color='gray', linestyle=':', linewidth=1.2)
axes[0].scatter([0.5], [1.0], color='crimson', s=80, zorder=5)
axes[0].annotate('$H_{\\max} = \\log_2(2) = 1$ bit', xy=(0.52, 1.01), fontsize=10)
axes[0].set_xlabel('Probability $p$', fontsize=12)
axes[0].set_ylabel('$H(p)$ [bits]', fontsize=12)
axes[0].set_title('Binary Shannon Entropy\n$H = -p\\log_2 p - (1-p)\\log_2(1-p)$', fontsize=11)
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(-0.05, 1.15)

# --- Right: maximum entropy H_max = log2(U) vs number of symbols U ---
U_values = np.arange(1, 33)
H_max_values = np.log2(np.maximum(U_values, 1))  # log2(1)=0

axes[1].plot(U_values, H_max_values, 'darkorange', linewidth=2.5, marker='o', markersize=5)
axes[1].set_xlabel('Number of distinct symbols $U$', fontsize=12)
axes[1].set_ylabel('$H_{\\max} = \\log_2(U)$ [bits]', fontsize=12)
axes[1].set_title('Maximum Entropy vs. Alphabet Size\n(uniform distribution)', fontsize=11)
axes[1].grid(True, alpha=0.3)
# Annotate a few key values
for u, h in [(2, 1.0), (4, 2.0), (8, 3.0), (16, 4.0), (32, 5.0)]:
    axes[1].annotate(f'U={u}: {h:.0f} bits', xy=(u, h), xytext=(u+0.5, h-0.3),
                     fontsize=8, color='darkred')

plt.tight_layout()
plt.savefig('shannon_entropy.png', dpi=100, bbox_inches='tight')
plt.show()
print("Figure saved.")
```

## Summary

### Key Formulas

| Concept | Formula |
|---|---|
| Information of a single event | $I(p) = \log_b\!\left(\dfrac{1}{p}\right) = -\log_b(p)$ |
| Change of base | $\log_b\!\left(\dfrac{1}{p}\right) = \dfrac{\log_a(1/p)}{\log_a(b)}$ |
| Shannon Entropy (discrete) | $H(\mathcal{X}) = -\displaystyle\sum_{j=1}^{U} p_j \log_b(p_j)$ |
| Shannon Entropy (continuous) | $H(X) = -\displaystyle\int p(x) \log_b p(x)\, dx$ |
| Total information | $I_{\text{total}} = N \cdot H(\mathcal{X})$ |
| Maximum entropy | $H_{\max} = \log_b(U)$, achieved when all $p_j = 1/U$ |

### Key Insights

1. **Rare events carry more information.** A surprise tells you more than the expected outcome.
2. **Certainty and impossibility carry no information.** $p=1$ and $p\to 0$ both give zero entropy contribution.
3. **The base of the logarithm sets the units** (bits, nats, trits), but carries no physical meaning.
4. **Entropy is maximized by uniform distributions.** Any bias toward predictable outcomes reduces entropy.
5. **Shannon entropy ≈ Boltzmann entropy** in structure — information theory and statistical mechanics share the same mathematical foundation.
